```python
# CLASSES and INSTANCES
'''In Python, a class is a blueprint for creating objects that have attributes (data) and methods (functions).it 
is a way to organize code and create reusable code components.'''
'''The __init__ method is a special method that is called when an instance of the class is created. It initializes
the attributes of the instance with the values passed as arguments.'''
'''he self is used to represent the instance of the class. With this keyword, you can access the attributes and methods of the class in python.'''

class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def bark(self):
        print(f"{self.name} barks!")

# creating an instance of the Dog class
my_dog = Dog("Rufus", 2)

# accessing attributes of the instance
print(f"My dog's name is {my_dog.name}.") # Output: "My dog's name is Rufus."
print(f"My dog is {my_dog.age} years old.") # Output: "My dog is 2 years old."

# calling methods of the instance
my_dog.bark() # Output: "Rufus barks!"
```

```python
#CLASS VARIABLES
'''In Python, class variables are variables that are defined at the class level and are 
shared among all instances of that class. This means that when the value of a class 
variable is changed, the change is reflected in all instances of the class.'''
class Car:
    wheels = 4   # this is a class variable
    def _init_(self, make, model):
        self.make = make
        self.model = model

my_car = Car("Toyota", "Corolla")
print(my_car.wheels)   # Output: 4

Car.wheels = 6   # changing the value of the class variable

another_car = Car("Honda", "Accord")
print(another_car.wheels)   # Output: 6
print(my_car.wheels)   # Output: 6
```

```python
# REGULAR METHOD, CLASS METHOD & STATIC METHOD
'''Regular methods in Python are the most common type of method and are defined inside a 
class. They are used to modify or access instance-specific state. Regular methods take self
 as the first parameter by convention.'''
class Car:
  def _init_(self, make, model, year):
    self.make = make
    self.model = model
    self.year = year
    self.odometer = 0

  def drive(self, miles):
    self.odometer += miles

my_car = Car("Toyota", "Corolla", 2020)
print(my_car.odometer)   # Output: 0

my_car.drive(50)
print(my_car.odometer)   # Output: 50


'''Class methods are methods that are bound to the class and not the instance of
 the class. They are used to modify or access class-specific state. Class methods
  take cls as the first parameter by convention.'''

class BankAccount:
    interest_rate = 0.02

    def _init_(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    @classmethod
    def set_interest_rate(cls, rate):
        cls.interest_rate = rate

account1 = BankAccount("1234", 5000)
account2 = BankAccount("5678", 8000)

print(account1.interest_rate)   # Output: 0.02
print(account2.interest_rate)   # Output: 0.02

BankAccount.set_interest_rate(0.03)

print(account1.interest_rate)   # Output: 0.03
print(account2.interest_rate)   # Output: 0.03

'''Static methods are methods that don't depend on the state of the instance or the
 class. They are defined inside a class but aren't bound to the instance or the class.
  They are used when a method doesn't depend on instance or class state but still belongs 
  logically to the class. Static methods don't take self or cls as a first parameter.'''

class Calculator:
    @staticmethod
    def add(x, y):
        return x + y

result = Calculator.add(2, 3)
print(result)   # Output: 5

```

```python
#INHERITANCE
'''Inheritance is a fundamental concept of object-oriented programming (OOP) that 
allows a class to inherit properties and methods from another class'''
# Define a Parent class
class Animal:
    def _init_(self, name):
        self.name = name
        
    def make_sound(self):
        print(f"{self.name} makes a sound.")

# Define a Child class that inherits from the Parent class
class Cat(Animal):
    def _init_(self, name):
        super()._init_(name) #The super()._init_() method initializes the attributes inherited from the parent class without having to rewrite the code to do so.
        
    def make_sound(self):
        print(f"{self.name} meows.")

# Create instances of both classes and call their methods
my_animal = Animal("Animal")
my_animal.make_sound()

my_cat = Cat("Garfield")
my_cat.make_sound()
```

```python
#SPECIAL METHODS
'''In Python, special methods, also known as magic methods, are predefined methods that
 define the behavior of built-in operations on objects of a class. These methods are always
  surrounded by double underscores (e.g., _init(), __str(),
 __add_(), etc.) and are automatically called when certain operations are performed on an object'''
 '''__init__(self[, argsâ€¦]): This method is called when an object is created and initialized.

__str__(self): This method returns a string representation of the object.

__repr__(self): This method returns a string that can be used to recreate the object.

__len__(self): This method returns the length of the object.

__add__(self, other): This method defines the behavior of the addition operator (+).

__sub__(self, other): This method defines the behavior of the subtraction operator (-).

__eq__(self, other): This method defines the behavior of the equality operator (==).'''
class Complex:
    def _init_(self, real, imag):
        self.real = real
        self.imag = imag

    def _str_(self):
        return f'{self.real} + {self.imag}i'

    def _add_(self, other):
        real = self.real + other.real
        imag = self.imag + other.imag
        return Complex(real, imag)

    def _eq_(self, other):
        return self.real == other.real and self.imag == other.imag

# Create two Complex objects
c1 = Complex(2, 3)
c2 = Complex(4, 5)

# Call the _str_() method
print(c1)    # Output: 2 + 3i

# Call the _add_() method
c3 = c1 + c2
print(c3)    # Output: 6 + 8i

# Call the _eq_() method
print(c1 == c2)    # Output: False
print(c2 == Complex(4, 5))    # Output: True
```
